```python
import pandas
```

```python
df = pandas.read_excel('./cleaned_time_series_data.xlsx')
```

```python
df.head()
```

```python
df.drop(['s.no', 'year', 'date'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('evp', axis=1)
y = df['evp']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=300, max_depth=30, random_state=42)
#  here n_estimators is the number of trees in the forest, max_depth is the maximum depth of the tree
# random_state is the seed used by the random number generator

# fit the model in the training data
rf.fit(X_train, y_train)
```

```python
# show tree
# from sklearn.tree import export_graphviz
# import pydot

# tree = rf.estimators_[5]
# export_graphviz(tree, out_file='tree.dot', feature_names=df.columns[:-1], rounded=True, precision=1)

# (graph, ) = pydot.graph_from_dot_file('tree.dot')
# graph.write_png('tree.png')
```

```python
rf.score(X_test, y_test)
```

```python
# predict
y_pred = rf.predict(X_test)

for i in range(len(y_pred)):
    print(f'Predicted: {y_pred[i]}, Actual: {y_test.iloc[i]}')
```
